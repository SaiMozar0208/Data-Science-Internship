```python
#BoxPlot-to find outliers(extreme distinct val than dataset)

marks=[27,23,25,81,9]
#outliers= 81,9 denoted by o
#orange line=median
#box=rnage of dataset
mp.boxplot(marks)
```

```python
q3=np.percentile(marks,75) #
q1=np.percentile(marks,25)
q3,q1
```

```python
iqr = q3-q1
iqr# interquartile range
```

```python
#Quartile deviation = (Q3-Q1)/2

#lowerbound & upperbound
lb=q1-1.5*iqr
ub=q3+1.5*iqr
lb,ub
```

```python
#SEABORN

tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib as mpp
mpp.rcParams.update(mpp.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='time')
mp.show()
```

```python
#histogram=freq distribution
mp.xlabel('tip amt')
mp.ylabel('freq')
sns.histplot(tips['tip'],kde=True) #kde is freq distribution curve
mp.show()
```

```python
#histogram=freq distribution
mp.xlabel('Day')
mp.ylabel('tip')
#sns.histplot(tips['tip'],kde=True)
sns.boxenplot(x='day',y='tip',data=tips)
mp.show()

#The boxen plot or letter value plot is similar to a box plot but provides more information about the shape of the distribution, particularly in the tails
```

```python
sns.pairplot(tips,hue='smoker')
mp.show()
#pairplot is similar to jointplot but
#1) scatter graph is generated for all the numeric values plkresent in the dataset
#2)diagonal graphs are the histogram
```
